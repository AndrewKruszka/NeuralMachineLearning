**2.3 Tutorial**

Perform the optimization problem of finding the minimum of J(θ) = (2θ-3)^2

**Part 1: Plot J(θ) vs θ**

To start, we will plot J(θ) vs θ. In order to do this, we need to first define J(θ) and set up the dependancies

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function J(theta)
def J(theta):
    return (2 * theta - 3) ** 2
```

Next we will plot the relationship between J(θ) and θ.

```python
# Generate values for theta
theta_values = np.linspace(-2, 4, 100)
J_values = J(theta_values)

# Plot J(theta) vs theta
plt.figure(figsize=(8, 5))
plt.plot(theta_values, J_values, label=r'$J(\theta) = (2\theta - 3)^2$', color='blue')
plt.xlabel(r'$\theta$')
plt.ylabel(r'$J(\theta)$')
plt.title('Plot of $J(\\theta) = (2\\theta - 3)^2$')
plt.axvline(x=1.5, color='red', linestyle='--', label=r'Optimal $\theta^* = 1.5$')
plt.legend()
plt.grid()
plt.show()
```

The chart above shows a parabolic relationship between J(θ) where the minimum occurs at θ* = 1.5

**Part 2: Determine the minimum using gradient descent**

Next, we will confirm that the minimum occurs at θ* = 1.5 by using gradient descent. We will use an initial value of θ = 5

To start, we need to find the derivative of the cost function J(θ), which can be accomplished by using the chain rule

```python
# Define the gradient of J(theta)
def gradient_J(theta):
    return 4 * (2 * theta - 3)
```

Next, we will implement the gradient descent algorithm: θ:=θ−(α⋅∇J(θ))

Where α is the learning rate (0.1) and ∇J(θ) is the gradient function

```python
# Implementing gradient descent
def gradient_descent(theta_init, alpha, num_iterations):
    theta = theta_init  # Initialize theta
    theta_history = [theta]  # Track theta values for visualization

    for i in range(num_iterations):
        theta = theta - alpha * gradient_J(theta)
        theta_history.append(theta)  # Store theta value

        # Print step-by-step updates
        print(f"Iteration {i+1}: theta = {theta:.5f}")

    return theta, theta_history
```

This algorithm works by plugging in the initial value of θ=5 into the gradient function then multiplying that value by the learning rate. You then subract that value from the initial θ value to determine a new θ value.

These steps are repeated until θ is reasonably close to 0. For the purposes of this tutorial, we will run 10 iterations.

```python
# Initial theta value
theta_init = 5
alpha = 0.1  # Learning rate
num_iterations = 10  # Perform 10 iterations

# Run gradient descent
theta_opt, theta_history = gradient_descent(theta_init, alpha, num_iterations)
```

The chart below shows how θ changes over time as we run through the gradient descent algorithm

```python
# Plot theta values over iterations
plt.figure(figsize=(8, 5))
plt.plot(range(len(theta_history)), theta_history, marker='o', linestyle='-', color='purple')
plt.xlabel('Iteration')
plt.ylabel(r'$\theta$ value')
plt.title('Convergence of $\theta$ in Gradient Descent')
plt.axhline(y=1.5, color='red', linestyle='--', label=r'Optimal $\theta^* = 1.5$')
plt.legend()
plt.grid()
plt.show()
```

As you can see in the chart, after about 5 iterations it's clear that the algorithm converges toward θ* = 1.5, which is where J(θ) reaches its minimum
